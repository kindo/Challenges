```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
from scipy import misc


data = misc.ascent()

plt.imshow(data)
```

```python
#Convolution

x_size = data.shape[0]
y_size= data.shape[1]

conv_filter = np.array([[10, 2, -10], [10, 5, -15], [15, 2, -8]])

x_conv = conv_filter.shape[0]

f = conv_filter.shape[0]

result = np.zeros(((x_size - f + 1),(x_size - f + 1) ))


for i in range(x_size - f+1):
    for j in range(y_size - f + 1):
        mat = data[i:i+x_conv, j:j+x_conv]
        conv_value = mat.flatten()@conv_filter.flatten()
        if conv_value < 0.:
            conv_value = 0.
        if conv_value > 225.:
            conv_value = 225.
        result[i, j] = conv_value
```

```python
cmap = plt.cm.inferno
plt.figure(figsize=(5, 5))
plt.imshow(result, cmap=cmap)
```

```python
#pooling 

x_size = result.shape[0]
y_size= result.shape[1]

p = 2

pooling = np.zeros((int(x_size/2), int(y_size/2)))

for i in range(0, x_size - p+1, 2 ):
    for j in range(0, y_size - p + 1, 2):
        mat = result[i:i+p, j:j+p]
        pooling[int(i/2), int(j/2)] = np.max(mat)
```

```python
cmap = plt.cm.inferno
plt.figure(figsize=(5, 5))
plt.imshow(pooling, cmap=cmap)
```
